```python
import cv2
import numpy as np
from scipy.linalg import svd
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("mei.jpeg",cv2.IMREAD_UNCHANGED)
```

```python
img_grey = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) 
```

```python
cv2.imwrite("mei_grey.bmp",img_grey)
```

```python
image = Image.open('./mei_grey.bmp') 
A = np.array(image)
plt.imshow(A, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

```python
# 对图像矩阵A进行奇异值分解，得到p,s,q
p,s,q = svd(A, full_matrices=False)
```

```python
# s_temp = np.zeros(s.shape[0])
# s.shape[0] * 1 // 100
```

```python
# 取前k个特征，对图像进行还原
def get_image_feature(s, k):
    # 对于S，只保留前百分之k的特征值
    s_temp = np.zeros(s.shape[0])
    n = s.shape[0] *k // 100
    s_temp[0:n] = s[0:n]
    s = s_temp * np.identity(s.shape[0])
    # 用新的s_temp，以及p,q重构A
    temp = np.dot(p,s)
    temp = np.dot(temp,q)
    plt.imshow(temp, cmap=plt.cm.gray, interpolation='nearest')
    plt.show()
    print(A-temp)
```

```python
get_image_feature(s, 1)
```

```python
get_image_feature(s, 10)
```

```python
get_image_feature(s, 50)
```
